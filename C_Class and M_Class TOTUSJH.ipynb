### First level decomposition 

```python
data = c_class_data
```

```python


def decompose_and_plot_wavelets(data, wavelet='sym4', level=1):
    """
    Plot wavelet coefficients for a given signal.

    Parameters:
    - data: Input signal
    - wavelet: Wavelet type (default is 'sym4')
    - level: Level of decomposition (default is 1)
    """
    coeffs = pywt.wavedec(data, wavelet=wavelet, level=level)

    plt.figure(figsize=(24, 15))

    for i in range(len(coeffs)):
        plt.subplot(len(coeffs), 1, i + 1)
        plt.plot(coeffs[i])

    plt.tight_layout()
    plt.show()
    return coeffs


```

```python
coeffs1 = decompose_and_plot_wavelets(data, wavelet='sym4', level=1)
```

### second level decomposition

```python
coeffs2 = decompose_and_plot_wavelets(data, wavelet='sym4', level=2)
```

### third level decomposition

```python
coeffs3 = decompose_and_plot_wavelets(data, wavelet='sym4', level=3)
```

### Approximation aka low frequency plots at different levels(1,2,3) for C_class flare

```python
plt.subplot(4,1, 1)
plt.plot(c_class_data, label='Original Signal', linewidth=2)
plt.title('original signal')
plt.subplot(4,1, 2)
plt.plot(coeffs1[0])
plt.title('level1')
plt.subplot(4,1, 3)
plt.plot(coeffs2[0])
plt.title('level2')
plt.subplot(4,1, 4)
plt.plot(coeffs3[0])
plt.title('level3')
```
